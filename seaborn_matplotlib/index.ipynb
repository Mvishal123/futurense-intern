```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv("./datasets/titanic_train.csv")

print(raw_data.dtypes)
```

```python
print(raw_data.head(3))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=raw_data, x='Age', bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Count plot of passengers by embarked port
plt.figure(figsize=(10, 6))
sns.countplot(data=raw_data, x='Embarked')
plt.title('Count of Passengers by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot of fare by passenger class and survival
plt.figure(figsize=(10, 6))
sns.boxplot(data=raw_data, x='Pclass', y='Fare', hue='Survived')
plt.title('Fare by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Basic line plot
plt.figure(figsize=(10, 6))
plt.plot(raw_data['Age'].head(10))
plt.title('Age of the First 10 Passengers')
plt.xticks(range(len(raw_data['Age'].head(10))))
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.show()
```

```python
# Scatter plot
colors = ["blue", "red"]

plt.figure(figsize=(10, 6))
plt.scatter(raw_data['Age'], raw_data['Fare'], c=colors[1])
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
pclass_counts = raw_data['Pclass'].value_counts()
plt.bar(pclass_counts.index, pclass_counts.values)
plt.title('Number of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(raw_data['Age'].dropna(), bins=30)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
plt.boxplot(raw_data['Fare'].dropna())
plt.title('Box Plot of Fares')
plt.xlabel('Fare')
plt.show()
```

```python
# Subplots
plt.figure(figsize=(14, 6))

# Subplot 1: Histogram of Age
plt.subplot(1, 2, 1)
# drop null values
plt.hist(raw_data['Age'].dropna(), bins=40) 
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')

# Subplot 2: Box plot of Fare
plt.subplot(1, 2, 2)
plt.boxplot(raw_data['Fare'].dropna())
plt.title('Box Plot of Fares')
plt.xlabel('Fare')

plt.tight_layout()
plt.show()
```

```python
# Save the scatter plot to a file
plt.figure(figsize=(10, 6))
plt.scatter(raw_data['Age'], raw_data['Fare'], color='green', alpha=0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.savefig('age_vs_fare.png')
plt.show()
```
