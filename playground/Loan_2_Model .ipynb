#Splitting Our data into Train , Test

Splitting the data to input and output

```python
y = df["Loan Sanction Amount (USD)"]
x = df.drop(columns=["Customer ID", "Name", "Loan Sanction Amount (USD)"])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the training data
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)

y_train_scaled = scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler.fit_transform(y_test.values.reshape(-1, 1))

# Convert scaled data back to DataFrame
x_train = pd.DataFrame(x_train_scaled, columns=x_train.columns)
y_train = pd.DataFrame(y_train_scaled, columns=['Loan Sanction Amount (USD)'])
x_test = pd.DataFrame(x_test_scaled, columns=x_test.columns)
y_test = pd.DataFrame(y_test_scaled, columns=['Loan Sanction Amount (USD)'])
```

#Load the Dataset

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self, data, label):
        self.data = torch.tensor(data.values).type(torch.float32)
        self.label = torch.tensor(label.values).type(torch.float32).view(-1, 1)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        return self.data[index], self.label[index]
```

```python
dataset = CustomDataset(x_train , y_train)
```

```python
batch_size = 2
train_loader = DataLoader(dataset , batch_size = batch_size , shuffle = True)
```

```python
class NeuralNetworks(nn.Module):
    def __init__(self, in_features, out_features, hidden_units):
        super().__init__()
        self.in_features = in_features  # Store input features
        self.layer = nn.Sequential(
            nn.Linear(in_features=in_features, out_features=hidden_units),
            nn.ReLU(),
            nn.Linear(in_features=hidden_units, out_features=hidden_units),
            nn.ReLU(),
            nn.Linear(in_features=hidden_units, out_features=out_features)
        )

    def forward(self, x):
        return self.layer(x)
```

```python
in_features = x_train.shape[1]
out_features = 1
print(in_features)
model = NeuralNetworks(in_features = in_features , out_features = out_features , hidden_units = 100)
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(params = model.parameters() , lr = 0.001)
```

Function to find Mean Absolute Error

```python
def mean_absolute_error(preds, targets):
    return torch.abs(preds - targets).mean()
```

Training the Model

```python
from tqdm.auto import tqdm
#Random Seed
torch.manual_seed(42)

epochs = 10

for epoch in tqdm(range(epochs)):
    model.train()
    epoch_loss = 0
    epoch_mae = 0
    for input, target in train_loader:
        # Forward pass
        train_logits = model(input)

        # Calculate the loss
        train_loss = loss_fn(train_logits, target)
        epoch_loss += train_loss.item()

        # Calculate MAE
        train_mae = mean_absolute_error(train_logits, target)
        epoch_mae += train_mae.item()

        # Optimizer steps
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()

    avg_epoch_loss = epoch_loss / len(train_loader)
    avg_epoch_mae = epoch_mae / len(train_loader)
    print(f'Epoch: {epoch+1}, Train Loss: {avg_epoch_loss:.4f}, Train MAE: {avg_epoch_mae:.4f}')

# Print final training loss and MAE
print(f'Final Train Loss: {avg_epoch_loss:.4f}, Final Train MAE: {avg_epoch_mae:.4f}')
```

Testing the test Dataset

```python

test_dataset = CustomDataset(x_test, y_test)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

model.eval()
test_loss = 0
test_mae = 0
with torch.no_grad():
    for input, target in test_loader:
        test_logits = model(input)
        loss = loss_fn(test_logits, target)
        mae = mean_absolute_error(test_logits, target)
        test_loss += loss.item()
        test_mae += mae.item()

avg_test_loss = test_loss / len(test_loader)
avg_test_mae = test_mae / len(test_loader)
print(f'Test Loss: {avg_test_loss:.4f}, Test MAE: {avg_test_mae:.4f}')
```
